```python
import pandas as pd

# creating a messy dataset
data = {
    "Crops": ["Maize", "Rice", "Bean", None, "Yam"],
    "Yield_kg": [1500, 2000, None, 1200, 3000],
    "Price_Naira": [350, 400, 800, 500, None]

}

df = pd.DataFrame(data)
print("Original messy data")
print(df)
```

```python
print("Missing values in each column")
print(df.isnull().sum())


print("\nTotal missing values:", df.isnull().sum().sum())
```

```python
# Fix missing crop name
df["Crops"] = df["Crops"].fillna("Unknown")

# Fix missing numbers with average
df["Yield_kg"] = df["Yield_kg"].fillna(df["Yield_kg"].mean())
df["Price_Naira"] = df["Price_Naira"].fillna(df["Price_Naira"].mean())

print("Cleaned data:")
print(df)
```

```python
# adding duplicate rows to our data
df_duplicates = df._append(df.iloc[0])
df_duplicates = df_duplicates._append(df.iloc[1])

print("Data with Duplicates:")
print(df_duplicates)
print("\nTotal rows:", len(df_duplicates))
```

```python
# remove duplicate rows
df_cleaned = df_duplicates.drop_duplicates()

print("Cleaned data:")
print(df_cleaned)
print("\nTotal rows:", len(df_cleaned))
```

```python
# rename columns
df_cleaned = df_cleaned.rename(columns={"Crops": "Crop", "Yield_kg": "Yield (kg)", "Price_Naira": "Price (Naira)"})

print("Data with renamed columns:")
print(df_cleaned)
```
